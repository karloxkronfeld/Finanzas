```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = yf.download("^GSPC", period="2y", interval="1h")[["Close"]]
```

```python
# Obtener los valores de fecha y precio
fecha = data.index
precio = data.Close
# Crear la figura y el objeto del eje
fig, ax = plt.subplots()
# Graficar los datos del S&P 500
ax.plot(fecha, precio, color='b')

# Definir las coordenadas del rectángulo
x = fecha[np.argmin(precio)]  # Coordenada x del vértice inferior izquierdo del rectángulo MINIMO GLOBAL
y = min(precio)  # Coordenada y del vértice inferior izquierdo del rectángulo MINIMO GLOBAL
width =  fecha[np.argmax(precio)]-x  # Ancho del rectángulo
print(fecha[0])
height = max(precio) - min(precio)  # Altura del rectángulo


# Añadir el rectángulo al gráfico
rect = plt.Rectangle((x, y), width, height, edgecolor='r', facecolor='none')
ax.add_patch(rect)

# print(x)
# print(y)
# print(f"width {width}")
# print(height)
```
