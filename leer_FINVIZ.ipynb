```python
import pandas as pd
import requests
import numpy as np
```

```python
def leer_pagina(url):
  header ={"User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.75 Safari/537.36","X-Requested-With": "XMLHttpRequest"}
  r=requests.get(url, headers=header)
  dfs=pd.read_html(r.text)
  dfs=dfs[-2]
  dfs.columns=dfs.loc[0]
  dfs=dfs[1:].iloc[:,1:]
  return dfs  
```

```python
semanal = leer_pagina("https://finviz.com/screener.ashx?v=141&f=ind_exchangetradedfund&o=-perf1w")
mensual = leer_pagina('https://finviz.com/screener.ashx?v=141&f=ind_exchangetradedfund&o=-perf4w')
quat =    leer_pagina("https://finviz.com/screener.ashx?v=141&f=ind_exchangetradedfund&o=-perf13w")
semestre= leer_pagina("https://finviz.com/screener.ashx?v=141&f=ind_exchangetradedfund&o=-perf26w")
anual   = leer_pagina("https://finviz.com/screener.ashx?v=141&f=ind_exchangetradedfund&o=-perf52w")
```

```python
periodos= [semanal,mensual,quat,semestre]
```
