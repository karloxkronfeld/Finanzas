```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=yf.download("^GSPC",period="2y",interval="1h").reset_index()
data.columns=["Fecha","Apertura","Maximo","Minimo","Cierre","Cierre_Adj","Volume"]
# data.index=data.Fecha
# data
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar los datos en un DataFrame de pandas


# Obtener los índices de los mínimos y máximos
min_indices = data['Cierre'].idxmin()
max_indices = data['Cierre'].idxmax()

# Obtener los valores de los mínimos y máximos
min_values = data['Cierre'].min()
max_values = data['Cierre'].max()

# Graficar los datos y la línea que une los mínimos y máximos
plt.figure(figsize=(10, 6))
plt.plot(data.index, data['Cierre'], label='Precio de cierre')
plt.scatter(min_indices, min_values, color='red', label='Mínimos')
plt.scatter(max_indices, max_values, color='green', label='Máximos')
plt.plot([min_indices, max_indices], [min_values, max_values], color='blue', linestyle='dashed', label='Línea de mínimos y máximos')
plt.xlabel('Fecha')
plt.ylabel('Precio')
plt.legend()
plt.title('Línea que une mínimos y máximos del S&P 500')
plt.show()
```
