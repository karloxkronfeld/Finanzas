```python
import pandas as pd
```

```python
import pandas_datareader as data
```

```python
qqq=data.DataReader("QQQ","yahoo","2020-01-01","2021-01-01")
spy=data.DataReader("SPY","yahoo","2020-01-01","2021-01-01")
```

```python
qqq["dif"]=qqq["Adj Close"].diff(1)
spy["dif"]=spy["Adj Close"].diff(1)
```

```python
spy.dif.plot()
qqq.dif.plot()
```

```python
datos=pd.DataFrame([qqq.dif,spy.dif],index=["qqq","spy"]).T
```

```python
datos.spy.plot()
```

```python
import numpy as np
```

```python
data=data.get_data_yahoo("NFLX","2020-01-01")
```

```python
high_low=data.High-data.Low
high_close=np.abs(data.High-data.Close.shift())
low_close=np.abs(data.Low-data.Close.shift())

ranges=pd.concat([high_low,high_close,low_close],axis=1)
true_range=np.max(ranges,axis=1)

ATR= true_range.rolling(14).sum()/14
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ATR.plot(ax=ax)
data['Close'].plot(ax=ax, secondary_y=True, alpha=0.3)
plt.show()
```
