```
import pandas_datareader as data
import time
from pylab import *
from matplotlib import animation,rc
from IPython.display import HTML
```

```
COP=data.DataReader("USDCOP","av-intraday","2021-03-05","2021-03-10",api_key="000W9HI05KWAWA7B")["close"].rename(index="COP")
EUR=data.DataReader("EURUSD","av-intraday","2021-03-05","2021-03-10",api_key="000W9HI05KWAWA7B")["close"]
JPY=data.DataReader("USDJPY","av-intraday","2021-03-05","2021-03-10",api_key="000W9HI05KWAWA7B")["close"]
```

```
COP.rename(index="Fecha")
```

```
lista=[EUR,COP,JPY]
lista_str=["EUR","COP","JPY"]
```

```
while True:
  for x,y in zip(lista,lista_str):
    time.sleep(4)
    print("\r {}:{:,.2f}".format(y,x),end="")
```

```
def race_tickers(ticker):
  ax.clear()
  seleccion=lista[ticker]
  ax.plot(seleccion,label=lista_str[ticker])
  
  legend()
  xticks([])
  
  
  
```

```
fig_race,ax=subplots()
race_tickers(2)
```

```
fig_race,ax=subplots()
frames=range(len(lista))
mi_anim=animation.FuncAnimation(fig_race,race_tickers,frames,interval=2000)
# rc("animation",ffmpeg_path="/usr/bin/ffmpeg")
HTML(mi_anim.to_html5_video())
```
