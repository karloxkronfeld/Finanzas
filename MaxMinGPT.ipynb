```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=yf.download("^GSPC",period="2y",interval="1h").reset_index()
data.columns=["Fecha","Apertura","Maximo","Minimo","Cierre","Cierre_Adj","Volume"]
data.index=data.Fecha
# data
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Cargar los datos en un DataFrame de pandas

# Dividir los datos en subconjuntos
num_parts = 5
data_parts = np.array_split(data, num_parts)

# Encontrar los mínimos de cada subconjunto
min_values = [part['Cierre'].min() for part in data_parts]

# Obtener los índices de los mínimos de cada subconjunto
min_indices = [part['Cierre'].idxmin() for part in data_parts]

# Graficar los datos originales
plt.figure(figsize=(10, 6))
plt.plot(data.index, data['Cierre'], label='Precio de cierre')

# Marcar los mínimos de cada subconjunto
for i, idx in enumerate(min_indices):
    plt.scatter(idx, min_values[i], color='red', label=f'Mínimo {i+1}')

# Unir los mínimos con una línea diagonal
plt.plot(min_indices, min_values, color='blue', linestyle='dashed', label='Unión de mínimos')

plt.xlabel('Fecha')
plt.ylabel('Precio')
plt.legend()
plt.title('Unión diagonal de mínimos de cada subconjunto del S&P 500')
plt.show()
```

```python


# Dividir los datos en subconjuntos
num_parts = 5
data_parts = np.array_split(data, num_parts)

# Encontrar los máximos de cada subconjunto
max_values = [part['Cierre'].max() for part in data_parts]

# Obtener los índices de los máximos de cada subconjunto
max_indices = [part['Cierre'].idxmax() for part in data_parts]

# Graficar los datos originales
plt.figure(figsize=(10, 6))
plt.plot(data.index, data['Cierre'], label='Precio de cierre')

# Marcar los máximos de cada subconjunto
for i, idx in enumerate(max_indices):
    plt.scatter(idx, max_values[i], color='green', label=f'Máximo {i+1}')

# Unir los máximos con una línea diagonal
plt.plot(max_indices, max_values, color='orange', linestyle='dashed', label='Unión de máximos')

plt.xlabel('Fecha')
plt.ylabel('Precio')
plt.legend()
plt.title('Unión diagonal de máximos de cada subconjunto del S&P 500')
plt.show()
```
