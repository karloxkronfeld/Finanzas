```python
import pandas as pd
from pandas_datareader import data 
from pylab import *
```

```python
datos=data.DataReader("EURUSD","av-intraday","2021-04-20","2021-04-24",api_key="000W9HI05KWAWA7B")
```

```python
close=datos.close
```

```python
close
```

```python
t=20  # selecciona los ultimos 20
factor_stdev=2
seleccion=[]
sma_values=[]
upper_band=[]
lower_band=[]

for x in close:
  seleccion.append(x)
  if len(seleccion)>t:
    del seleccion[0]    

  
  upper_band.append(x+(x*0.001))
  lower_band.append(x-(x*0.001))

```

```python
dataframe= pd.DataFrame(columns=["close","upper","lower"])
dataframe["close"]=close
dataframe["upper"]=upper_band
dataframe["lower"]=lower_band
dataframe.plot(figsize=(16,6))
```

```python
tiempo=200
datos["up"]=datos["high"].rolling(window=tiempo).max()
datos["down"]=datos["low"].rolling(window=tiempo).min()
datos["mid"]=(datos.down+datos.up)/2
datos[["close","up","down","mid"]].plot(figsize=(15,5))
```
