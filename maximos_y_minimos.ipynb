```python
from alpha_vantage.timeseries import TimeSeries
import pandas as pd
from pylab import *
```

```python
time_series=TimeSeries(key="000W9HI05KWAWA7B",output_format="pandas")
data=time_series.get_intraday("EURUSD",interval="1min",outputsize="full")
```

```python
data=data[0][::-1].iloc[:,[3]]
data.columns=["close"]
```

```python
temporalidades=[]
for x in range(1,13):
  temporalidades.append(data.iloc[-60*x:])
```

```python
horas_maximos=[]
for x in temporalidades:
  horas_maximos.append(x.idxmax().close) 
```

```python
precios_maximos=[]
for x in temporalidades:
  precios_maximos.append(x.max())
```

```python
figure(figsize=(18,7))
plot(data.close)
for i in range(0,11):
  plot([horas_maximos[i],horas_maximos[i+1]],[precios_maximos[i],precios_maximos[i+1]],color="r")
```
