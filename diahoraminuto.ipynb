```python
import pandas as pd
from pylab import *
```

```python
datos=pd.read_csv("GOLD-HastaMarzo15.xlsx")
datos=datos.iloc[:,[1,2,3]]
datos["precio"]=(datos.bid+datos.ask)/2
datos=datos.groupby("time").mean()
datos.index=pd.DatetimeIndex(datos.index)
datos.index=datos.index.strftime("%Y-%m-%d %H:%M")
minuto=datos.groupby("time").mean()
```

```python
datos.index=pd.DatetimeIndex(datos.index)
datos.index=datos.index.strftime("%Y-%m-%d %H")
hora=datos.groupby("time").mean()
```

```python
datos.index=pd.DatetimeIndex(datos.index)
datos.index=datos.index.strftime("%Y-%m-%d")
diaria=datos.groupby("time").mean()
```

```python
subplots(0,0)
minuto.precio.plot()
title("minuto")
subplots(0,0)
hora.precio.plot()
title("hora")
subplots(0,0)
diaria.precio.plot()
title("diaria")
```
